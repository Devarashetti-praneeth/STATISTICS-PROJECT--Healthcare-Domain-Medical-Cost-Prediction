# Treatment of Outliers Using IQR in Bmi Column

```python
q1 = data["bmi"].quantile(0.25)
q3 = data["bmi"].quantile(0.75)
Iqr = q3-q1
bmi_lower =q1-1.5*Iqr
bmi_upper = q3+1.5*Iqr
print("q1 = ",q1)
print("q3 =",q3)
print("Iqr =",Iqr)
print("bmi_lower = ",bmi_lower)
print("bmi_upper = ",bmi_upper)
```

# Cleaned Data

```python
 clean_data = data[(data["bmi"]< bmi_upper)]
 
```
